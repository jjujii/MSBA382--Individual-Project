```python
import pandas as pd

# Load dataset
df = pd.read_csv("healthcare-dataset-stroke-data.csv")

# Step 0: Drop the single 'Other' gender row
df = df[df['gender'].isin(['Male','Female'])]

# Step 1: Handle missing BMI with group-wise median
df['age_group'] = pd.cut(df['age'], bins=[0,18,35,55,100],
                         labels=['Child','Young Adult','Middle Age','Senior'])
df['bmi'] = df.groupby(['gender','age_group'])['bmi'].transform(
    lambda x: x.fillna(x.median())
)

# Step 2: Encode categorical variables
df['gender'] = df['gender'].map({'Male':0, 'Female':1})
df['ever_married'] = df['ever_married'].map({'No':0, 'Yes':1})
df['Residence_type'] = df['Residence_type'].map({'Rural':0, 'Urban':1})

# One-hot encode work_type and smoking_status (keep Unknown explicitly)
df = pd.get_dummies(df, columns=['work_type','smoking_status'], drop_first=False)

# Step 3: Feature engineering
df['BMI_category'] = pd.cut(df['bmi'], bins=[0,18.5,25,30,100],
                            labels=['Underweight','Normal','Overweight','Obese'])
df = pd.get_dummies(df, columns=['BMI_category'], drop_first=False)

# Step 4: Finalize prepared dataframe
prepared_df = df.drop(columns=['id'])

# Save cleaned version
prepared_df.to_csv("stroke_data_cleaned.csv", index=False)
```
